5) Подсчёт оценок студентов (на 12)

Создайте функцию analyze_grades(grades_list),
которая принимает список оценок студентов (от 0 до 100)
и возвращает словарь с подсчётом:

"passed" — количество студентов, у которых оценка ≥ 60

"failed" — количество студентов, у которых оценка < 60

"average" — средняя оценка группы

Пример:

grades = [85, 42, 77, 90, 58, 63, 100]
print(analyze_grades(grades))


Результат:

{'passed': 5, 'failed': 2, 'average': 73.57}

```python
def analyze_grades(grades_list):
    grade_counts = {}
    passed = 0
    failed = 0
    for i in grades_list:
        if i >=60:
            passed +=1
        if i <60:
            failed +=1
    avarege=sum(grades_list)/len(grades_list)
    grade_counts["passed"] = passed
    grade_counts["failed"] = failed
    grade_counts["avarege"] = (round (avarege, 2))
    return grade_counts
        
grades = [85, 42, 77, 90, 58, 63, 100]
print(analyze_grades(grades)) 
```
